```julia
using JuMP, Ipopt, Test
```

```julia
function example_basic(; verbose = true)
    model = Model(with_optimizer(Ipopt.Optimizer))

    @variable(model, 0 <= x <= 2)
    @variable(model, 0 <= y <= 30)

    @objective(model, Max, 5x + 3y)
    @constraint(model, 1x + 5y <= 3.0)

    if verbose
        print(model)
    end

    JuMP.optimize!(model)

    obj_value = JuMP.objective_value(model)
    x_value = JuMP.value(x)
    y_value = JuMP.value(y)

    if verbose
        println("Objective value: ", obj_value)
        println("x = ", x_value)
        println("y = ", y_value)
    end

    @test obj_value ≈ 10.6
    @test x_value ≈ 2
    @test y_value ≈ 0.2
end

example_basic(verbose = false)
```

# Vector example

```julia
function example_basic(; verbose = true)
    model = Model(with_optimizer(Ipopt.Optimizer))

    @variable(model, 0 <= x[1:2] <= 2)

    @objective(model, Max, 5x[1] + 3x[2])
    @constraint(model, 1x[1] + 5x[2] <= 3.0)

    if verbose
        print(model)
    end

    JuMP.optimize!(model)

    obj_value = JuMP.objective_value(model)
    x_value = JuMP.value.(x)

    if verbose
        println("Objective value: ", obj_value)
        println("x = ", x_value)
    end

    @test obj_value ≈ 10.6
    @test x_value ≈ [2, 0.2]
end

example_basic(verbose = true)
```

# Nonlinear example

```julia
model = Model(with_optimizer(Ipopt.Optimizer))
@variable(model, x, start = 0.0)
@variable(model, y, start = 0.0)

@NLobjective(model, Min, (1 - x)^2 + 100 * (y - x^2)^2)

optimize!(model)
println("x = ", value(x), " y = ", value(y))

# adding a (linear) constraint
@constraint(model, x + y == 10)
optimize!(model)
println("x = ", value(x), " y = ", value(y))
```
